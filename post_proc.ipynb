```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
from IPython.display import HTML
import sys
```

```python
f = np.loadtxt("settings.txt", dtype=int)
```

```python
rows = f[0]
cols = f[1]
p = f[2]
p_rows = f[3]
p_cols = f[4]
steps = f[5]

p_data = [] # each iteration one after the other in the same file
sub_row = [] # each processor sub rows
sub_col = [] # each processor sub cols
for i in range(p):
    output = "output_" + str(i) + ".txt"
    sub_set = "sub_set_" + str(i) + ".txt"
    p_data.append(np.loadtxt(output, dtype=int))
    sub_row.append(np.loadtxt(sub_set, dtype=int)[0])
    sub_col.append(np.loadtxt(sub_set, dtype=int)[1])
```

```python
p_size = p_rows * p_cols
p_mod = p_size // p_rows
offset = 0
grid = []
for period in range(steps):
    for mod in range(0, p, p_mod): # find how many rows to append for 1 iteration
        tmp_board = np.empty(cols)
        for row_iter in range(sub_row[mod]): # append amount of rows in one proc
            tmp_row = np.array([])
            for i in range(p_cols):
                tmp_row = np.concatenate([tmp_row, p_data[i][row_iter+period*(sub_row[mod]-1),:]])
            tmp_board = np.vstack([tmp_board, tmp_row]) 
    tmp_board = np.delete(tmp_board, 0, 0)
    grid.append(tmp_board)
```

```python
fig = plt.figure(figsize=(10,10))
imgs = []
for frames in range(period):
    plt.title('Game of Life')
    img = plt.imshow(grid[frames], cmap='gray')
    imgs.append([img])
ani = anim.ArtistAnimation(fig, imgs, interval=100, blit=True)

plt.close(fig)   # prevent final frame plot from showing up inline below
HTML(ani.to_jshtml())
```
